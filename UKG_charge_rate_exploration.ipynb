```python
params={"page":0,"per_Page":50}
```

```python
all_results = []
```

```python
while True:
    params["page"] += 1
    r = sesh.get(endpoint_url,params=params)
    current_page = r.json()
    if not current_page:
        break
    all_results.extend(current_page)
```
